# Prediction using Supervised ML

```python
import pandas as pd
data = pd.read_csv("http://bit.ly/w-data")
print("Data imported successfully")
#Print first five row of our dataset
data.head()
#Print data columns name in our dataset
data.columns
#Chech whether is there any null value present in our
# dataset
data.isnull().sum()
data.info()
#From above we can observe that there is no null value in our dataset so there is no need of cleaning the data
```

```python
#imorting required libraries..
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data.plot(x='Hours', y='Scores', style='o')    
plt.title('Hours vs Percentage')    
plt.xlabel('The Hours Studied')    
plt.ylabel('The Percentage Score')    
plt.show()
```

```python
X = data.iloc[:, :-1].values    
y = data.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split    
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression    
model = LinearRegression()    
model.fit(X_train, y_train)   
  
print("Training ... Completed !.")
print(model.intercept_)
print(model.coef_)
# Plotting the regression line
line = model.coef_ * X+ model.intercept_  

# Plotting for the test data

plt.scatter(X, y)
plt.plot(X, line,'m')
plt.title("Plotting the regression Line")
plt.ylabel('Scores')
plt.xlabel('Hours')
plt.show()
```

```python
#print(x_test) # Testing data - In Hours
y_pred = model.predict(X_test) # Predicting the scores
y_pred
data = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
data
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,model.predict(X_train),color='m')
plt.title('Training Set')
plt.xlabel('Hours')
plt.ylabel('Scores');
```
